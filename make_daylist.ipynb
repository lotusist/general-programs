```python
def make_daylist(fromdate:str, todate:str):
   if len(fromdate) != 8 or len(todate) !=8 :
      print("inputs must be 8 digits. ex) '20221231'")
      return None  
   elif int(todate) < int(fromdate):
      print("todate must be later than fromdate.")
      return None
   elif todate[:4] == fromdate[:4]:
      _from_month = int(fromdate[4:6])
      _to_month = int(todate[4:6])
      iter_num = (_to_month - _from_month)-1

      if iter_num > 0:
         day_list = list()
         i = int(fromdate)
         while i % 100 <= 31:
            day_list.append(str(i))
            i += 1


         for _ in range(0, iter_num):
            i += 69
            while i % 100 <= 31:
               day_list.append(str(i))
               i += 1
         i +=69

         for _ in range(i, int(todate)+1):
            day_list.append(str(i))
            i += 1

      elif iter_num == 0:
         day_list = list()
         i = int(fromdate)
         while i % 100 <= 31:
            day_list.append(str(i))
            i += 1
         i += 69
         
         for _ in range(i, int(todate)+1):
            day_list.append(str(i))
            i += 1
      
      elif iter_num == -1:
         day_list = list()
         for j in range(int(fromdate), int(todate)+1):
            day_list.append(j)

   elif int(todate[:4]) > int(fromdate[:4]):
      year_diff = int(todate[:4]) - int(fromdate[:4])
      day_list = list()
      for year in range(0, year_diff):
            day_list = day_list + make_daylist(fromdate, str(int(fromdate[:4])+year)+'1231')
      day_list = day_list + make_daylist(todate[:4]+'0101', todate)
   
      year_range = list()
      for y in range(int(fromdate[:4]), int(todate[:4])+1):
         year_range.append(str(y))
      
      ab_list = list()
      ab_dates = ['0229', '0230', '0231', '0431', '0631', '0931', '1131']
      for year_element in year_range:
         for date_element in ab_dates:
            ab_list.append(year_element + date_element)
      
      for ab in ab_list:
         if ab not in day_list:
            pass
         else:
            day_list.remove(ab)

   return day_list
```
